```python
# Importing Data
import pandas as pd
df=pd.read_csv(r"C:\Users\hp\Downloads\customer_shopping_behavior.csv")
df.head()
```

```python
# Checking the summry statistic of  numeric data..
df.describe()
```

```python
# Checking for Null Columns in the Data
df.isnull().sum()
```

```python
# Filling the NUll Columns with Median depending on their Cateogry
df["Review Rating"]=df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
# Changing Columns Name into Smaller case and snake casing insted of Space" "..
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.columns
```

```python
# Creat a Column age_group
labels=["Young Adult","Adult","Middle-Aged","Seniors"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=labels)
df[["age","age_group"]].head()
```

```python
# Create column Purchase_frequecy_days
frequency_mapping = {
    "Fortnightly":14,
    "Weekly":7,
    "Quartly":90,
    "Bi-Weekly":14,
    "Annually":365,
    "Every 3 Months":90
}
df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["frequency_of_purchases","purchase_frequency_days"]].head()
```

```python
# Now Next we will check "discount_applied" and "promo_code_used" is same or not
(df["discount_applied"]==df["promo_code_used"]).all()
```
